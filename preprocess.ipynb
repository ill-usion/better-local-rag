```python
import os
import pandas as pd
from pathlib import Path

print(pd.__version__)
```

```python
if not os.path.isdir("./data"):
    os.mkdir("./data")
```

```python
sheets = ["Mining", "Building", "Farming", "Animals", "Monsters", "Weaponry", "Biomes", "Adventure", "Redstone", "Enchanting", "Statistics", "Nether", "Potions", "End", "Super Challenges"]
include_tab_name = True
include_parent_col = True
include_actual_req_col = True
include_hidden_col = True
include_item_rewards_col = False
include_xp_rewards_col = False
include_trophy_col = False
include_source_col = False
include_version_added_col = False
include_notes_col = True
```

```python
paths = []
for category in sheets:
    print(category)

    filtered = pd.DataFrame()
    sheet = pd.read_excel("BlazeandCave's Advancements Pack 1.20 documentation.xlsx", sheet_name=category)
    # Separate advancements from advancement requirements
    advs = sheet[~(sheet["Parent"].isna() & sheet["Hidden?"].isna() & sheet["Item rewards"].isna() & sheet["XP Rewards"].isna() & sheet["Trophy"].isna() & sheet["Source"].isna())]
    reqs = sheet[sheet["Parent"].isna() & sheet["Hidden?"].isna() & sheet["Item rewards"].isna() & sheet["XP Rewards"].isna() & sheet["Trophy"].isna() & sheet["Source"].isna()]
    reqs = reqs.dropna(thresh=2)

    # Fill in the advancement requirements
    for (adv, desc, *_) in reqs.itertuples(index=False, name=None):
        try:
            advs.loc[advs["Advancement Name"].str.contains(adv[:-1]if len(adv) else adv), "Actual Requirements (if different)"] = desc
        except:
            print(f"Failed at {adv}: {desc}")

    filtered["Advancement Name"] = advs["Advancement Name"]
    filtered["Description"] = advs["Description"]

    if include_tab_name:
        filtered["Tab"] = category
    if include_parent_col:
        filtered["Parent"] = advs["Parent"]
    if include_actual_req_col:
        filtered["Actual Requirements"] = advs["Actual Requirements (if different)"] 
    if include_hidden_col:
        filtered["Hidden?"] = advs["Hidden?"]
    if include_item_rewards_col:
        filtered["Item Rewards"] = advs["Item rewards"]
    if include_xp_rewards_col:
        filtered["XP Rewards"] = advs["XP Rewards"]
    if include_trophy_col:
        filtered["Trophy"] = advs["Trophy"]
    if include_source_col:
        filtered["Source"] = advs["Source"]
    if include_version_added_col:
        filtered["Version Added"] = advs["Version added"]
    if include_notes_col:
        filtered["Notes"] = advs["Notes"]

    path = f"./data/{category}.csv"
    paths.append(path)
    filtered.to_csv(path)
```

```python
for p in paths:
    print(Path(p).resolve())
```
